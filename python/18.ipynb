```python
import numpy as np
import networkx as nx
```

```python
with open("../data/18.txt") as f:
    data = [line.rstrip().split(",")[::-1] for line in f.readlines()]

byts = []
for d in data:
    byts.append((int(d[0]),int(d[1])))
n = 7 if len(byts)<500 else 71
nc = nr = n
drops = 12 if len(byts)<500 else 1024
```

```python
def get_idx(r,c,nc=nc):
    return (r*nc)+c

def invert_idx(idx,nc,nr):
    return (idx//nr,idx%nc)


directions = [
    (-1,0), 
    (1,0),
    (0,1),
    (0,-1)
]
```

```python
grid = np.zeros((nr,nc))
for byt in byts[:drops]:
    grid[*byt] = 1

def take_step(i,step,grid=grid):
    nr,nc = grid.shape
    move = (i[0]+step[0],i[1]+step[1])
    if move[0]>=0 and move[1]>=0 and move[0]<nc and move[1]<nr and grid[move]!=1:
        return move    

# turn the grid into a graph
g = nx.Graph()
i=0
for r in range(nr):
    for c in range(nc):
        assert i == get_idx(r,c,nc)
        g.add_node(i)
        i+=1

# add edges
for r in range(nr):
    for c in range(nc):
        if grid[r,c] != 1:
            steps = [take_step((r,c),d,grid) for d in directions ]
            for step in steps:
                if step:
                    g.add_edge(get_idx(r,c),get_idx(*step))
```

```python
if drops ==12:
    t= 6
else:
    t=70

p=nx.shortest_path(g,source=0,target=get_idx(t,t,nc))
len(p)-1
```

```python
# part two
```

```python
for byt in byts[drops:]:
    g.remove_node(get_idx(*byt,nc))
    if not nx.has_path(g,source=0,target=get_idx(t,t,nc)):
        print(byt[::-1])
        break
```

```python
output = ""
for r in range(nr):
    for c in range(nc):
        idx = get_idx(r,c,nc)
        if idx in p:
            output += "O"
        elif grid[r,c] ==0:
            output += "."
        elif grid[r,c] ==1:
            output += "#"
    output += "\n"
```
